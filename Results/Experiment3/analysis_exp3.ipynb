```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

# Experiment 3

```python
df = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'], dtype="float")

df2 = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'], dtype="float")

df_ = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'], dtype="float")

df2_ = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'], dtype="float")
```

```python
dfa = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'])

dfa2 = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'])

dfa_ = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'])

dfa2_ = pd.DataFrame(columns=['0', '20', '40', '60', '80', '100'], 
                  index=['100', '80', '60', '40', '20', '0'])
```

```python
import os

for file in os.listdir():
    filename = os.fsdecode(file)
    if 'peaks' in filename:
        file_split = filename.split('.')[0].split('_')
        jobID = file_split[-3]
        if int(jobID) <= 36:
            col = file_split[-1]
            row = file_split[-2]
            df1 = pd.read_csv(filename)
            d1 = df1.peaks/33700 * 100
            d2 = df1.locaitons
            df[col][row] = np.mean(d1)
            df2[col][row] = np.mean(d2)
        else:
            col = file_split[-1]
            row = file_split[-2]
            df1 = pd.read_csv(filename)
            d1 = df1.peaks/33700 * 100
            d2 = df1.locaitons
            df_[col][row] = np.mean(d1)
            df2_[col][row] = np.mean(d2)
```

```python
for file in os.listdir():
    filename = os.fsdecode(file)
    if 'peaks' in filename:
        file_split = filename.split('.')[0].split('_')
        jobID = file_split[-3]
        if int(jobID) <= 36:
            col = file_split[-1]
            row = file_split[-2]
            df1 = pd.read_csv(filename)
            d1 = df1.peaks/33700 * 100
            d2 = df1.locaitons
            dfa[col][row] = d1
            dfa2[col][row] = d2
        else:
            col = file_split[-1]
            row = file_split[-2]
            df1 = pd.read_csv(filename)
            d1 = df1.peaks/33700 * 100
            d2 = df1.locaitons
            dfa_[col][row] = d1
            dfa2_[col][row] = d2
```

```python
s = sns.heatmap(df, annot=True, fmt='.3g',
               cbar_kws={'label': 'Infectios Peak Size (%)'})
s.set_ylabel('Proportion Locked Down (%)', fontsize=12)
s.set_xlabel('Proportion Wearing Masks (%)', fontsize=12)
s.set_title('P(MT) = 0.33', fontsize=12, weight='bold')
```

```python
s = sns.heatmap(df2, annot=True, fmt='.3g', cmap="viridis", 
               cbar_kws={'label': 'Infectios Peak Location (Day)'})
s.set_ylabel('Proportion Locked Down (%)', fontsize=12)
s.set_xlabel('Proportion Wearing Masks (%)', fontsize=12)
s.set_title('P(MT) = 0.33', fontsize=12, weight='bold')
```
